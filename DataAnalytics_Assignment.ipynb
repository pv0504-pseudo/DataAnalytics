```python
#list of numbers from 1 to 10 and print their respective ordinal
for i in range(1,11):
  if i == 1:
    print(f"{i}st")
  elif i == 2:
    print(f"{i}nd")
  elif i == 3:
    print(f"{i}rd")
  else:
    print(f"{i}th")
```

```python
#Function to check if the number is odd or even
def check_even_odd(num):
     if num % 2 == 0:
        return "Even"
     else:
        return "Odd"
# Input from the user
num = int(input ("Enter a number:"))
#Display the result
result = check_even_odd(num)
print(f"The number {num} is {result}.")
#Loop two times
for i in range(2):
    num = int(input("Enter a number:"))
    result = check_even_odd(num)
    print(f" The number {num} is {result}.")



```
